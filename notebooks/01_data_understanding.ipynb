# **Churn Analysis - Waze App**

## Step 01 - Data Understanding


### imports and data loading

```python
import pandas as pd
import numpy as np

df =  pd.read_csv('waze_dataset.csv')
```

### summary information

```python
print(df.head(10))
print()
print(df.info())
print()
print(df.describe())
```

**Conclusion:** <br>
The dataset has 700 missing values in the 'label' column.

We also recognize that there is the possibility of outlier values in many columns (We will investigate that later).

### investigate the 'label' column

```python
null_df = df[df['label'].isnull()]

null_df.describe()
```

```python
not_null_df = df[~df['label'].isnull()]

not_null_df.describe()
```

**Conclusion:** <br>
The mean and standart deviations are fairly consistent between the missing 'label' and non-missing 'label' data. <br>
Comparing the summary stats of missing and non-missing 'label' data reveals nothing remarkable.

Check if there is a connection between missing label data and the 'device' variable

```python
print(null_df['device'].value_counts())
print()
print(null_df['device'].value_counts(normalize=True))
```

Compare to the device ratio in the full dataset to check if there is a non-random cause of the missing data

```python
print(df['device'].value_counts())
print()
print(df['device'].value_counts(normalize=True))
```

**Conclusion:** <br>
The percentage of missing values by each device is consistent with their represantation of the data overall.

There is nothing to suggest a non-random cause of the missing data.

### churned vs. retained users

```python
print(df['label'].value_counts())
print()
print(df['label'].value_counts(normalize=True))
```

**Conclusion:** <br>
This dataset contains 82% retained users and 18% churned users.

#### Calculate Median

Next, we calculate the median values of all columns of churned and retained users. <br>
The reason for calculating the median instead of the mean value is that we don't want outliers to unduly affect the potrayal of a typical user.

```python
df.groupby('label').median(numeric_only=True)
```

**Conclusion:** <br>
Users who churned drove averaged ~3 more drives in the last month than retained users, but retained users used the app on over twice as many days as churned users in the same time period.

The median churned user drove ~200 more kilometers and 2.5 more hours during the last month than the median retained user.

It seems that churned users had more drives in fewer days, and their trips were farther and longer in duration. Perhaps this is suggestive of a user profile.

#### Investigate distance

We calculate the median kilometers per drive in the last month for both retained and churned users.

```python
# Add a column to df called `km_per_drive`
df['km_per_drive'] = df['driven_km_drives'] / df['drives']

# Group by `label`, calculate the median, and isolate for km per drive
median_km_per_drive = df.groupby('label').median(numeric_only=True)[['km_per_drive']]
median_km_per_drive
```

**Conclusion:** <br>
The median retained user drove about one more kilometer per drive than the median churned user.

How many kilometers per driving day was this?

```python
# Add a column to df called `km_per_driving_day`
df['km_per_driving_day'] = df['driven_km_drives'] / df['driving_days']

# Group by `label`, calculate the median, and isolate for km per driving day
median_km_per_driving_day = df.groupby('label').median(numeric_only=True)[['km_per_driving_day']]
median_km_per_driving_day
```

```python
# Add a column to df called `drives_per_driving_day`
df['drives_per_driving_day'] = df['drives'] / df['driving_days']

# Group by `label`, calculate the median, and isolate for drives per driving day
median_drives_per_driving_day = df.groupby('label').median(numeric_only=True)[['drives_per_driving_day']]
median_drives_per_driving_day
```

**Conclusion:** <br>
The median user who churned drove 698 kilometers each day they drove last month, which is ~240% the per-drive-day distance of retained users. The median churned user had a similarly disproporionate number of drives per drive day compared to retained users.

### Conclusion


It is clear from these figures that, regardless of whether a user churned or not, the users represented in this data are serious drivers! It would probably be safe to assume that this data does not represent typical drivers at large.

Perhaps the data and in particular the sample of churned users contains a high proportions of long-haul truckers.

It's possible that the reason for their driving so much is also the reason why the Waze app does not meet their specific set of needs, which may differ from the needs of a more typical driver, such as a commuter.
